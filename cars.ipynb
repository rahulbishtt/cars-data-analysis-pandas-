Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column.


```python
car.isnull().sum()
```

```python
car
```

```python
car.fillna(car.mean(numeric_only=True),inplace=True)
```

```python
for col in car.select_dtypes(include=['object']).columns:
    car[col].fillna(car[col].mode()[0], inplace=True)
```

```python
car.isnull().sum()
```
